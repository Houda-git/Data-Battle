```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Charge ton fichier
df = pd.read_csv("segment_alerts_all_airports_train.csv")

# Garde uniquement les distances valides
df = df[df["dist"].notna()].copy()

# Bornes de 5 km
max_dist = np.ceil(df["dist"].max() / 5) * 5
bins = np.arange(0, max_dist + 5, 5)

# Histogramme
plt.figure(figsize=(10, 6))
plt.hist(df["dist"], bins=bins, edgecolor="black")

plt.xlabel("Distance (km)")
plt.ylabel("Nombre d'éclairs")
plt.title("Répartition des éclairs par intervalles de 5 km")
plt.xticks(bins)
plt.grid(axis="y", alpha=0.3)

plt.show()
```
