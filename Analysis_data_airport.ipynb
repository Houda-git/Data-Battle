### Data Inspection Summary

The dataset contains **507,071 lightning events** recorded around **five airports** and described by **13 variables** including temporal (`date`), spatial (`lon`, `lat`), physical lightning characteristics (`amplitude`, `maxis`, `icloud`), and airport-related information (`dist`, `azimuth`, `airport`).  
Each row represents a single lightning strike detected within a **30 km radius around an airport**.  
The variables have different data types, including numerical variables (`float`, `int`), a categorical variable (`airport`), and a boolean variable (`icloud`).  
The `date` variable is currently stored as a string and will need to be converted to a **datetime format** to enable temporal analysis.  
Similarly, `airport_alert_id` is stored as a float due to the presence of missing values but will later be converted to a **nullable integer type (`Int64`)** since it represents an identifier.  
Most variables contain **no missing values**, except for `airport_alert_id` and `is_last_lightning_cloud_ground`, which contain about **89% missing values**.  
These missing values are expected because, according to the documentation, these variables are only recorded for lightning events occurring within **20 km of an airport**.  
Lightning events are unevenly distributed across the five airports, with **Pise, Bastia, and Biarritz** recording the highest number of observations.  
The amplitude values show a wide range of **positive and negative intensities**, corresponding to different lightning polarities, while the localization error (`maxis`) remains relatively small, indicating precise lightning detection.  
Finally, the variable `is_last_lightning_cloud_ground` identifies whether a lightning strike corresponds to the **last cloud-to-ground lightning of an alert sequence**, making it particularly relevant for the project objective of **predicting the end of lightning events**.

### Data Cleaning 

#### Handling missing values

The variables `airport_alert_id` and `is_last_lightning_cloud_ground` contain many missing values because they are only recorded for lightning events occurring within 20 km of an airport.  
For modeling purposes, `airport_alert_id` is converted to a nullable integer type (`Int64`), while missing values in `is_last_lightning_cloud_ground` are replaced with `False`, since lightning events outside the alert zone cannot correspond to the last lightning of an alert.

```python
df["airport_alert_id"] = df["airport_alert_id"].astype("Int64")
df["is_last_lightning_cloud_ground"] = df["is_last_lightning_cloud_ground"].fillna(False)
df["is_last_lightning_cloud_ground"] = df["is_last_lightning_cloud_ground"].astype("bool")


df["date"] = pd.to_datetime(df["date"])

df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["hour"] = df["date"].dt.hour
df["minute"] = df["date"].dt.minute

df[["date","year","month","day","hour"]].head()
```

```python
alerts = (
    df[df["airport_alert_id"].notna()]
    .groupby(["airport","airport_alert_id"])
    .agg(
        start_time=("date","min"),
        end_time=("date","max"),
        n_lightnings=("lightning_id","count"),
        nb_cloud_ground=("icloud", lambda x: (x==False).sum()),
        nb_intra_cloud=("icloud", lambda x: (x==True).sum()),
        max_distance=("dist","max"),
        mean_amplitude=("amplitude","mean")
    )
)

df_ml = df.drop(columns=["is_last_lightning_cloud_ground"])

alerts["duration_min"] = (
    alerts["end_time"] - alerts["start_time"]
).dt.total_seconds()/60

alerts[["duration_min"]].head()
```

```python
# df doit contenir 'airport', 'airport_alert_id', 'date', 'amplitude', 'dist', 'icloud'
df["date"] = pd.to_datetime(df["date"])

# créer table pour ML
features_list = []

for alert_id, alert_df in df[df["airport_alert_id"].notna()].groupby(["airport","airport_alert_id"]):
    alert_df = alert_df.sort_values("date")
    alert_df["cum_lightnings"] = range(1, len(alert_df)+1)
    alert_df["cum_cloud_ground"] = (alert_df["icloud"]==False).cumsum()
    alert_df["cum_intra_cloud"] = (alert_df["icloud"]==True).cumsum()
    alert_df["cum_amplitude"] = alert_df["amplitude"].cumsum()
    alert_df["elapsed_min"] = (alert_df["date"] - alert_df["date"].iloc[0]).dt.total_seconds()/60
    alert_df["cum_max_distance"] = alert_df["dist"].cummax()
    
    features_list.append(alert_df)

df_features = pd.concat(features_list)
```
